```python
import numpy as np
data_NotMNIST = np.load('/home/xaMars/wjr_py/l2pJittor/l2p-jittor-main/output/task_list_arr.npy')
```

```python
data_NotMNIST=data_NotMNIST.reshape(10,-1,3)
```

```python
data_NotMNIST.shape
```

```python
# 2
import matplotlib.pyplot as plt
import numpy as np

steps = np.arange(data_NotMNIST.shape[1])

for task_id in range(10):
    # 
    fig, ax = plt.subplots(figsize=(12, 6))
    ax2 = ax.twinx()
    
    # 
    loss_values = data_NotMNIST[task_id, :, 0]
    acc_values = data_NotMNIST[task_id, :, 1]
    
    # loss (y)
    line1 = ax.plot(steps, loss_values, 'b-', linewidth=2, label='Loss')
    ax.set_xlabel('Training Steps', fontsize=12)
    ax.set_ylabel('Loss', color='b', fontsize=12)
    ax.tick_params(axis='y', labelcolor='b')
    
    # acc@1 (y)
    line2 = ax2.plot(steps, acc_values, 'r-', linewidth=2, label='Acc@1')
    ax2.set_ylabel('Accuracy@1', color='r', fontsize=12)
    ax2.tick_params(axis='y', labelcolor='r')
    
    # 
    ax.set_title(f'Task {task_id+1}: Loss and Accuracy@1 Curves', 
                fontsize=14, fontweight='bold', pad=15)
    
    # 
    ax.grid(True, alpha=0.3)
    
    # 
    lines1, labels1 = ax.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
    
    plt.tight_layout()
    plt.show()
    
    # 
    # plt.savefig(f'NotMNIST_task_{task_id+1}.png', dpi=300, bbox_inches='tight')

```
